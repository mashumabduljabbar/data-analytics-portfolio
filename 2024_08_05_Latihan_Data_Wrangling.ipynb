# **Data Wrangling**

Proses pengolahan data, atau yang dikenal sebagai data wrangling, melibatkan serangkaian aktivitas yang mencakup pengumpulan data, penilaian data, serta pembersihan data. Tahap-tahap ini dilakukan sebelum data dapat digunakan dalam analisis lebih lanjut. Pengumpulan data (gathering data) adalah tahap awal di mana data dikumpulkan dari berbagai sumber. Setelah itu, data dinilai atau dievaluasi (assessing data) untuk memastikan kualitas dan relevansinya. Terakhir, data dibersihkan (cleaning data) dari kesalahan atau ketidakkonsistenan agar siap digunakan dalam proses analisis.

# 4.Cleaning Data

Apabila pada proses sebelumnya kita menemukan masalah (missing value, outlier, dll.) yang terdapat di dalam sebuah data, masalah tersebut harus dibersihkan sebelum masuk tahap analisis data.

## 4.1 Membersihkan Data customers_df

### 4.1.1 Menghilangkan duplicate data

```python
customers_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi: ", customers_df.duplicated().sum())
```

### 4.1.2 Menangani missing value

```python
customers_df[customers_df.gender.isna()]
```

```python
customers_df.gender.value_counts()
```

```python
customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
customers_df.isna().sum()
```

### 4.1.3 Menangani inaccurate value

```python
customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
customers_df[customers_df.age == customers_df.age.max()]
```

```python
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
```

```python
customers_df.describe()
```

## 4.2 Membersihkan Data orders_df

### 4.2.1 Mengganti Tipe Data

```python
datetime_columns = ["order_date", "delivery_date"]

for column in datetime_columns:
  orders_df[column] = pd.to_datetime(orders_df[column])
```

```python
orders_df.info()
```

## 4.3 Membersihkan Data product_df

### 4.3.1 Menghilangkan duplicate data

```python
product_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah duplikasi: ", product_df.duplicated().sum())
```

## 4.4 Membersihkan Data sales_df

### 4.4.1 Menangani missing value

```python
sales_df[sales_df.total_price.isna()]
```

```python
sales_df["total_price"] = sales_df["price_per_unit"] * sales_df["quantity"]
```

```python
sales_df.isna().sum()
```
